```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1, 1, 100)

y = x**2 * (x + 1)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y, label=r'$y = x^2 (x + 1)$')
ax.fill_between(x, y, where=(y >= 0), color='green', alpha=0.5, label=r'$y \geq 0$')
ax.axhline(0, color='black')
ax.axvline(0, color='black')
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$y$')
ax.legend()
ax.grid()

plt.tight_layout()
plt.show()
```

```python
def f(x1,x2):
    return -(x1 - 1)**2 - (x2 - 2)**2
x1 = np.linspace(-2, 4, 100)
x2 = np.linspace(-2, 4, 100)
X1, X2 = np.meshgrid(x1, x2)
Z = f(X1, X2)
fig = plt.figure()
ax = fig.add_subplot(111)
contour = ax.contourf(X1, X2, Z, levels=20)
plt.clabel(contour, inline=True, fontsize=8)
ax.contourf(X1, X2, Z, levels=20, cmap='viridis')
ax.set_xlabel(r'$x_1$')
ax.set_ylabel(r'$x_2$')
ax.set_title(r'$f(x_1, x_2) = -(x_1 - 1)^2 - (x_2 - 2)^2$')

cb = plt.colorbar(contour, ax=ax)
cb.set_label(r'$f(x_1, x_2)$')
cb.ax.tick_params(labelsize=8)

plt.tight_layout()
plt.show()
```
