```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt

def f(x,y):
    return x**2 - 2*x + (x*y)**2 - 2*x*y

x0, xf = -100, 100
y0, yf = -100, 100

x = np.linspace(x0, xf, 100)
y = np.linspace(y0, yf, 100)

X, Y = np.meshgrid(x,y)

Z = f(X, Y)

fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')
ax.scatter(1,1, f(1,1), color='red', s=100,label='Minimum')
d = ax.plot_surface(X,Y, Z, cmap='inferno')

cb = plt.colorbar(d)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x,y)')
fig.legend()
plt.show()
```
